```python
def get_holidays(year=None):
    if year is None:
        return None
    
    year = str(year)
    URL = "https://www.timeanddate.com/holidays/philippines/" + year
    page = requests.get(URL)
    soup = BeautifulSoup(page.content, "html.parser")
    
    date = []
    day = []
    name = []
    holiday_type = []
    
    table = soup.find(id="holidays-table")
    rows = table.findChildren(["tr"])
    
    for row in rows:
        date_ = row.find("th", attrs={"class": "nw"})
        if date_ is None:
            continue
        date_ = date_.text

        day_ = row.find("td", attrs={"class": "nw"})
        if day_ is None:
            continue
        day_ = day_.text

        name_ = row.find("a")
        if name_ is None:
            continue
        name_ = name_.text

        holiday_type_ = row.find_all("td")
        if holiday_type_ is None:
            continue
        holiday_type_ = holiday_type_[-2].text

        date.append(date_)
        name.append(name_)
        day.append(day_)
        holiday_type.append(holiday_type_)
        
    return date, day, name, holiday_type
```
